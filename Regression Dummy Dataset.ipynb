#Step 2: Generating Dataset

```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=500, n_features=5, coef=False, bias=12, noise=10, random_state=2529)
```

```python
X, y, z = make_regression(n_samples=500, n_features=5, coef=True, bias=12, noise=10, random_state=2529)
```

#Step 3: Splitting Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state=2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

#Step 4: Creating model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

#Step 5: Training Model

```python
model.fit(X_train, y_train)
```

#Step 6: Getting Intercepts and Coefficients

```python
model.intercept_
```

```python
model.coef_
```

#Step 7: Prediction Model

```python
y_pred = model.predict(X_test)
```

```python
y_pred.shape
```

```python
y_pred
```

#Step 8: Accuracy

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
mean_squared_error(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```
