#**Splitting Data**

```python
X=df.iloc[:,0:1].values
y=df.iloc[:,1].values
```

```python
X
```

```python
y
```

#**Creating Model**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

#**Training Model**

```python
lr.fit(X,y)
```

```python
lr.predict([[3.2]])
```

```python
lr.coef_ #slope m
```

```python
lr.intercept_ #c
```

```python
y_pred=lr.predict(X)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y,y_pred)
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred,color='r')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mean_squared_error(y,y_pred)
```

```python
mean_absolute_error(y,y_pred)
```
