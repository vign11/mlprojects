#**Missing Values**

```python
d={'Cricket':[5,2,4,np.nan,np.nan,5,3,2],
   'Baseball':[2,np.nan,3,np.nan,4,5,2,4],
   'Tennis':[3.0,4,2,5,4,3,5,2]}
df=pd.DataFrame(d)
df
```

#Handling Missing Values
@1. Drop the rows which have missing values

@2. Drop the columns with missing values

@3. Fill with zero

@4. Fill the missing values with mean/median/mode

@5. Fill using advanced methods

```python
df.dropna(axis=0)
```

```python
df.dropna(axis=1)
```

```python
df.fillna(value=0)
```

```python
df['Cricket'].fillna(value=df['Cricket'].mean(),inplace=True)
df
```

```python
df['Baseball'].fillna(value=df['Baseball'].median(),inplace=True)
df
```

```python
df['Tennis'].fillna(value=df['Tennis'].mode(),inplace=True)
df
```
