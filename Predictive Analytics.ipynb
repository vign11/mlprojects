#Step 1: Import library

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

#Step 2: Import Data

```python
df=pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/MPG.csv')
```

```python
df
```

```python
df.nunique()
```

#__Data Preprocessing__

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

```python
df.shape
```

```python
df.columns
```

#Auxillary Step: Remove Missing Value Rows

```python
df=df.dropna()
```

```python
df.info()
```

#__Data Visualisation__

```python
sns.pairplot(df,x_vars=['displacement', 'horsepower', 'weight','acceleration','mpg'])
```

```python
sns.regplot(x='displacement',y='mpg',data=df)
```

#Step 3: Define y and X

```python
y=df['mpg']
```

```python
y.shape
```

```python
X=df[['displacement', 'horsepower', 'weight','acceleration']]
```

```python
X.shape
```

#Auxillary Step: Data Scaling 

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
X=s.fit_transform(X)
```

```python
pd.DataFrame(X).describe()
```

#Step 4: Splitting Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=2529)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

#Step 5: Create a Model

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

#Step 6: Train a Model

```python
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

#Step 7: Predict Model

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

#Step 8: Accuracy

```python
from sklearn.metrics import mean_absolute_percentage_error,mean_absolute_error,r2_score
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

#__Polynomial Regression__

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
p=PolynomialFeatures(degree=2,interaction_only=True,include_bias=False)
```

```python
X_train1=p.fit_transform(X_train)
```

```python
X_test1=p.fit_transform(X_test)
```

```python
model.fit(X_train1,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred_p=model.predict(X_test1)
```

#Accuracy

```python
from sklearn.metrics import mean_absolute_percentage_error,mean_absolute_error,r2_score
```

```python
mean_absolute_percentage_error(y_test,y_pred_p)
```

```python
mean_absolute_error(y_test,y_pred_p)
```

```python
r2_score(y_test,y_pred_p)
```
