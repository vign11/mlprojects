#__3. Artificial Nueral Network__

```python
import pandas as pd
```

```python
data = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Diabetes.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
y=data['diabetes']
```

```python
X=data[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age']]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X=sc.fit_transform(X)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

if we have 3 layers

```python
model=Sequential()
model.add(Dense(9, input_dim=8, activation='relu'))
model.add(Dense(12,  activation='tanh'))
model.add(Dense(10,  activation='relu'))
model.add(Dense(1,  activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=['accuracy'])
```

```python
model.fit(X,y,epochs=100,batch_size=100)
```
