**Guidelines for the k(no. of clusters) value in clustering**


*  The number of clusters should be square root of the total number of data points
*  We can use ELBOW method to find the optimal number of clusters


```python
from sklearn.cluster import KMeans
k=range(2,14) #since k value is sqrt of total data points that is 200
SSE=[]
for i in k:
  km=KMeans(n_clusters=i,random_state=13)
  km.fit(x)
  SSE.append(km.inertia_)
plt.plot(k,SSE)
plt.show()
```

We can see graph is having an elbow at 5 

```python
k=5
```

```python
model=KMeans(n_clusters=k,random_state=13)
```

```python
model.fit(x)
```

```python
y=model.predict(x)
y
```
